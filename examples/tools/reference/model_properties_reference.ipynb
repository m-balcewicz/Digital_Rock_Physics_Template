# Model Properties Reference
Show quick extraction of model properties for a small synthetic raw file using `get_model_properties`.

```python
import os, numpy as np, tempfile
from drp_template.tools import get_model_properties

# Create a temporary raw file with uint8 data (20x20x10)
nx, ny, nz = 20, 20, 10
data = np.random.randint(0, 4, (nx*ny*nz,), dtype=np.uint8)
tmp_raw = 'temp_model.raw'
data.tofile(tmp_raw)
props = get_model_properties(tmp_raw, dimensions={'nx': nx, 'ny': ny, 'nz': nz}, verbose=True)
print('Returned keys:', list(props.keys()))
```
