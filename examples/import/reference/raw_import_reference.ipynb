# Raw Import Reference
Minimal example of importing a raw binary file using `drp_template.io.import_model`.

This notebook assumes the working directory is this notebook's folder; data path is relative to `examples/data`.

```python
import os
from drp_template.io import import_model

# Path to example data (50x50x50 uint16)
data_path = os.path.abspath(os.path.join('..', '..', 'data', 'data_raw_uint16_50cube.raw'))
print('Data path:', data_path)

# Provide dimensions explicitly to avoid inference
dims = {'nx': 50, 'ny': 50, 'nz': 50}
model = import_model(file_path=data_path, dtype='uint16', voxel_size=5.0, dimensions=dims, order='C', paramsfile='raw_50cube.json')
print('Imported shape:', model.shape)
```
