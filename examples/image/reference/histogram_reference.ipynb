# Histogram Reference

This notebook shows how to use `drp_template.image.plotting.histogram` for grayscale intensity distributions and threshold highlighting.

```python
import numpy as np
import matplotlib.pyplot as plt
from drp_template.image.plotting import histogram

# Synthetic 3D grayscale volume with two modes
rng = np.random.default_rng(0)
a = rng.normal(60, 8, size=(120,120,120))
b = rng.normal(180, 10, size=(120,120,120))
mask = rng.random((120,120,120)) > 0.6
vol = a
vol[mask] = b[mask]
vol = np.clip(vol, 0, 255).astype(np.uint8)

# Define example thresholds
thresholds = [
    { 'label': 'Low', 'range': (0, 90) },
    { 'label': 'Mid', 'range': (90, 150) },
    { 'label': 'High', 'range': (150, 255) },
]

fig, ax = histogram(vol, thresholds=thresholds, dark_mode=True, title='Synthetic Intensity Histogram')
plt.show()
```
