This example is for importing regular and irregular 3D.raw files.

```python
# Import necessary libraries
import numpy as np
import barbara.input_output as io
from barbara.drp_template import plot_slice
```

First, start with the import of the raw data. During the import an automated binary check is applied to ensure the minimum of the data is 0. A
notification will be displayed.

```python
# Import regular cube
im = io.import_model(filename='raw_100cube.raw', dtype='uint16', shape=100)
```

After the import is successful the visualization of the central slice can be applied.

```python
fig = plot_slice(im, slice=0)
```

```python
im = io.import_model(filename='pore_100_100_400.raw', dtype='uint8', shape=(400, 100, 100))
fig = plot_slice(im, slice=0)
```

```python
# Import irregular data
im = io.import_model(filename='pore_50_100_400.raw', dtype='uint8', shape=(400, 100, 50))
# Convert the memmap array to a regular NumPy array
im_numpy = np.array(im)

fig = plot_slice(im_numpy)
```

Tiff sequences can also be imported. There the `filepath` to the directory is needed.

```python
directory = './tiff_sequences/pore_100_100_400'
dtype = 'uint8'
model = io.import_tiff_sequence(directory=directory, filename='pore_100_100_400', dtype=dtype)

fig = plot_slice(model)
```

It is also possible to import a raw XRCT image.

```python
directory = './tiff_sequences/raw_100_100_400'
dtype = 'uint16'
model = io.import_tiff_sequence(directory=directory, filename='raw_100_100_400', dtype=dtype)

fig = plot_slice(model, slice=0)
```

Or 3D tiff files

```python
# Import irregular data
im = io.import_tif_model(filename='3D_raw_100_100_200.tif')
# Convert the memmap array to a regular NumPy array
im_numpy = np.array(im)

fig = plot_slice(im_numpy, slice=0)
```
