# Binary Model Reference
Demonstrates inspecting a small synthetic binary (segmented) volume and summarizing unique values.

```python
import numpy as np
from drp_template.tools import get_value_statistics, classify_data_type

# Create synthetic binary data (0=background, 1=solid)
data = np.zeros((40, 30, 20), dtype=np.uint8)
data[10:30, 5:25, 3:15] = 1
stats = get_value_statistics(data)
dtype_class, phase_count = classify_data_type(stats['num_unique'])
print('Classification:', dtype_class)
print('Phase count:', phase_count)
print('Value counts:', stats['value_counts'])
```
