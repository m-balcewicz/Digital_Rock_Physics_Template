## Porous Media and Components

Definition: Porous media refer to materials that contain void spaces or pores. These void spaces can be filled with fluids (liquid or gas). Examples include soil, rocks, and biological tissues.

Characteristics:

Heterogeneous: Porous media often exhibit variations in structure and composition.
Permeability: The ability of a porous medium to transmit fluids.
Porosity: The fraction of void space in the total volume of the porous medium.

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Section 1.1: Porous Media
## Define characteristics of a simple porous medium
heterogeneous = True
permeability = 0.001
porosity = 0.2

## Print characteristics
print("### Section 1.1: Porous Media ###")
print(f"\n1. Porous Media Characteristics:")
print(f"Heterogeneous: {heterogeneous}")
print(f"Permeability: {permeability}")
print(f"Porosity: {porosity}")
```

## Components of Poroelastic Materials

Solid Matrix:

Definition: The solid framework or structure of the porous medium.
Properties: Elastic modulus, shear modulus, etc.
Deformation: The solid matrix can deform under stress.
Pore Fluids:

Definition: The fluids (liquid or gas) occupying the void spaces in the porous medium.
Properties: Density, viscosity, etc.
Interaction: Pore fluids interact with the solid matrix.

```python
# Section 2.1: Solid Matrix
## Define properties of the solid matrix
elastic_modulus = 1e6
shear_modulus = 5e5

## Print properties
print("\n### Section 2: Components of Poroelastic Materials ###")
print("\n2.1 Solid Matrix:")
print(f"- Elastic Modulus: {elastic_modulus} Pa")
print(f"- Shear Modulus: {shear_modulus} Pa")

# Section 2.2: Pore Fluids
## Define properties of pore fluids
fluid_density = 1000
fluid_viscosity = 0.001

## Print properties
print("\n2.2 Pore Fluids:")
print(f"- Fluid Density: {fluid_density} kg/m^3")
print(f"- Fluid Viscosity: {fluid_viscosity} Pa.s")
```

 ## Interaction Between Solid Matrix and Pore Fluids

Deformation of Solid Matrix:

When subjected to external stress, the solid matrix undergoes deformation.
This deformation can be elastic (recoverable) or inelastic (permanent).
Changes in Pore Pressure:

Deformation of the solid matrix leads to changes in pore pressure within the pore fluids.
Compression of the solid matrix increases pore pressure, and dilation decreases it.
Now, let's create a simple Python script to visualize the deformation of a solid matrix and the changes in pore pressure over time:

```python
# Section 3.1: Deformation of Solid Matrix
## Simulate solid matrix deformation over time
time = np.linspace(0, 10, 100)
solid_deformation = 0.1 * np.sin(2 * np.pi * time)

## Plot solid matrix deformation
plt.plot(time, solid_deformation)
plt.title('Section 3.1: Solid Matrix Deformation over Time')
plt.xlabel('Time')
plt.ylabel('Deformation')
plt.show()

# Section 3.2: Changes in Pore Pressure
## Simulate changes in pore pressure over time
pore_pressure = 0.05 * np.cos(2 * np.pi * time)

## Plot changes in pore pressure
plt.plot(time, pore_pressure)
plt.title('Section 3.2: Changes in Pore Pressure over Time')
plt.xlabel('Time')
plt.ylabel('Pore Pressure')
plt.show()
```
